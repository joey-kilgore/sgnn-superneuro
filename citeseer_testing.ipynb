```python
import sys
from superneuromat.neuromorphicmodel import NeuromorphicModel
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt


model = NeuromorphicModel()
```

```python
topics = []
paper_to_topic = {}
index_to_paper = []
    
f = open("data/citeseer/citeseer.content", 'r')
lines = f.readlines()
f.close()

for line in lines:
    fields = line.strip().split()
    if (fields[-1] not in topics):
        topics.append(fields[-1])
    paper_to_topic[fields[0]] = fields[-1]
    index_to_paper.append(fields[0])
train_papers = []
test_papers = []
```

```python
f = open("data/citeseer/citeseer_train_nodes.txt", 'r')
lines = f.readlines()
f.close()

check_breakdown = {}

for k in topics:
    check_breakdown[k] = 0
```

```python
while (len(train_papers) < len(topics)*20):
    index = np.random.randint(len(index_to_paper))
    topic = paper_to_topic[index_to_paper[index]]
    if (check_breakdown[topic] < 20 and index_to_paper[index] not in train_papers):
        train_papers.append(index_to_paper[index])
        check_breakdown[topic] += 1

for i in range(len(index_to_paper)):
    if (index_to_paper[i] not in train_papers):
        test_papers.append(index_to_paper[i])
```
